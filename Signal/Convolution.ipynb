```python
import matplotlib.pyplot as plt
import numpy as np

def normal_dist(x,sigma,miu):
    return (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-0.5*((x-miu)/sigma)**2)

x_data = np.linspace(-8, 8, 1000)

plt.figure(1, dpi=120)
plt.title('Normal Distribution Graph')
plt.plot(x_data,normal_dist(x_data,1,-2),label=r"$ \sigma = 1 , \mu = -2$")
plt.plot(x_data,normal_dist(x_data,1,-1),label=r"$ \sigma = 1 , \mu = -1$")
plt.plot(x_data,normal_dist(x_data,1,0),label=r"$ \sigma = 1 , \mu = 0$")
plt.plot(x_data,normal_dist(x_data,1,1),label=r"$ \sigma = 1 , \mu = 1$")
plt.plot(x_data,normal_dist(x_data,1,2),label=r"$ \sigma = 1 , \mu = 2$")
plt.legend()

plt.figure(2, dpi=120)
plt.title('Normal Distribution Graph')
# plt.plot(x_data,normal_dist(x_data,-2,0),label=r"$ \sigma = -2 , \mu = 0$")
plt.plot(x_data,normal_dist(x_data,-1,0),label=r"$ \sigma = -1 , \mu = 0$")
# plt.plot(x_data,normal_dist(x_data,0,0),label=r"$ \sigma = 0 , \mu = 0$") # not defined for zero
plt.plot(x_data,normal_dist(x_data,0.5,0),label=r"$ \sigma = 0.5 , \mu = 0$")
plt.plot(x_data,normal_dist(x_data,1,0),label=r"$ \sigma = 1 , \mu = 0$")
plt.plot(x_data,normal_dist(x_data,2,0),label=r"$ \sigma = 2 , \mu = 0$")
plt.plot(x_data,normal_dist(x_data,3,0),label=r"$ \sigma = 3 , \mu = 0$")
plt.plot(x_data,normal_dist(x_data,4,0),label=r"$ \sigma = 4 , \mu = 0$")
plt.legend()

plt.show()
```
