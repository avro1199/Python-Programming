```python
import numpy as np
import matplotlib.pyplot as plt

def f_2(x):
    return x**2

def f_3(x):
    return x**3

e = lambda x:np.exp(x)

x_data = np.linspace(-2,5,1000)
y_2_data = f_2(x_data)
y_3_data = f_3(x_data)

plt.figure(1,dpi=120)
plt.title('Plot of Exponential Functions')
plt.plot(x_data,y_2_data,'--',label='x$^2$')
plt.plot(x_data,y_3_data,'-.',label='x$^3$')
plt.plot(x_data,np.exp(x_data),'-.',label=r'$e ^x$')
plt.plot(x_data,0*x_data,'k')
plt.xlabel('X')
plt.ylabel('f(x)',rotation=0)
plt.grid(True)
plt.legend()
plt.show()
```
