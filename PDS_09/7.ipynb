```python
#fibonacchi series

num = int(input("Enter the Value of N: "))

#using recursion
#Method 1
fibo_nums = [-1]*num
fibo_nums[0] = 0
fibo_nums[1] = 1

def fibo_fnc(n):
    if fibo_nums[n] >= 0:
        return fibo_nums[n]
    fibo_nums[n] = fibo_fnc(n-1) + fibo_fnc(n-2)
    return fibo_nums[n]

# #Method 2
# def fib(n):
#     if n == 0:
#         return 0
#     if n == 1:
#         return 1
#     return fib(n-1) + fib(n-2)

for i in range(num):
    print(fibo_fnc(i), end = " ")
```

```python
#sum using recursion
num = int(input("Enter a number: "))

def sum_digits(n):
    if n == 0:
        return 0
    return n%10 + sum_digits(n//10)

print(f"Sum of all digits: {sum_digits(num)}")
```
