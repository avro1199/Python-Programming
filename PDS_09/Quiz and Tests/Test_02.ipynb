```python
#Assignment 01
#Line & Scatter Plots

import matplotlib.pyplot as plt
import numpy as np

#Line Plot
time = [6, 8, 10, 12, 14, 16, 18, 20, 22]
temp = [15, 18, 22, 26, 29, 31, 28, 24, 20]

plt.plot(time, temp, color='red', marker='*', linestyle='dashed', linewidth=2, markersize=12)
plt.title('Temperature vs Time')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend(['Line'])
plt.grid(True)
plt.show()

#Scatter Plot
plt.scatter(time, temp, color='blue', marker='x')
plt.title('Temperature vs Time')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend(['Scatter $^{Plot} \omega$'])
plt.grid(True)
plt.show()
```

```python
#Assignment 02
#Bar Chart and Pie Chart

#Bar Chart
brands = ['Apple', 'Samsung', 'Xiaomi', 'OnePlus', 'Vivo']
sales = [350, 420, 290, 180, 260]

plt.bar(brands, sales, color=['red', 'blue', 'green', 'yellow', 'orange'])
plt.title('Sales of Smartphones')
plt.xlabel('Brands')
plt.ylabel('Sales')
plt.grid(True)
plt.show()

#Pie Chart
plt.pie(sales, labels=brands, autopct='%1.2f%%', startangle=90,
        colors=['red', 'blue', 'green', 'yellow', 'orange'], explode=(0, 0.1, 0, 0, 0))

plt.title('Sales of Smartphones')
plt.show()
```

```python
#Assignment 03
#Histogram and Subplots

#Histogram
np.random.seed(42)
test_scores = np.random.normal(loc=70, scale=10, size=100)

hist_out = plt.hist(test_scores, bins=10, color='purple')
plt.axvline(np.mean(test_scores), color='red', linestyle='dashed', linewidth=2)
plt.axvline(np.median(test_scores), color='blue', linestyle='-.', linewidth=2)

plt.title('Test Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()

#Subplots
fig, axs = plt.subplots(2, 1)
axs[0].hist(test_scores, color='red')
axs[0].set_title('Test Scores')
axs[0].set_xlabel('Scores')
axs[0].set_ylabel('Frequency')
axs[0].grid(True)


#Cumulative Sum
cum_sum = np.cumsum(hist_out[0])

# axs[1].plot(cum_sum, hist_out[1][:-1], color='blue', marker='o')
axs[1].plot(hist_out[1][:-1], cum_sum, color='blue', marker='o')
axs[1].set_title('Cumulative Sum')
axs[1].set_xlabel('Scores')
axs[1].set_ylabel('Cumulative Sum')
axs[1].grid(True)
fig.tight_layout()

print(hist_out)

```
