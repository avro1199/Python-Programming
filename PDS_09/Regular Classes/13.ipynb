```python
import pandas as pd

titanic_df = pd.read_csv('DataSets\\train.csv')

titanic_df.info()
titanic_df
# titanic_df.shape

# titanic_df['Embarked'].unique()
# titanic_df['Cabin'].unique()
# titanic_df.notnull().sum()
```

```python
#Add title Mr. Mrs. Miss. Dr. Col.

titanic_df['Title'] = titanic_df['Name'].str.extract(r' ([A-Za-z]*)\.')

titanic_df
```

```python
titanic_df.groupby('Title').count()['Age']
```

```python
all_titles = titanic_df['Title'].unique()

title_map = {}
for title in all_titles:
    if(title == 'Mr'):
        title_map[title] = 0
    elif(title == 'Mrs'):
        title_map[title] = 1
    elif(title == 'Miss'):
        title_map[title] = 2
    else:
        title_map[title] = 3

print(title_map)
```

```python
titanic_df['Title'] = titanic_df['Title'].map(title_map)

titanic_df
```

```python
titanic_df['Age'].fillna(titanic_df.groupby('Title')['Age'].transform('median'), inplace=True)

titanic_df.info()
titanic_df
```
