```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2;

def g(x):
    return np.sin(x)

x_data = np.linspace(-30, 30, 1000)

plt.figure(1, dpi=240)
plt.title('Basic plotting of f(x) & sin(x)')
plt.plot(x_data,x_data*0, 'k--')
plt.plot(x_data*0,x_data, 'k--')
plt.plot(x_data,f(x_data), label='f(x)')
plt.plot(x_data,g(x_data), label='g(x)')
plt.plot(x_data,(g(x_data)*f(x_data)), label='f(x)*g(x)')
plt.plot(x_data,-f(x_data), label='-f(x)')

plt.xlim(-30, 30)
# plt.ylim(-6, 6)

plt.legend()
plt.show()
```
