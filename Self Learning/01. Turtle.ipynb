```python
import turtle

# Create a Turtle object
plane = turtle.Turtle()
plane.speed(2)  # Set the drawing speed

# Draw the body of the airplane
plane.penup()
plane.goto(-100, 0)
plane.pendown()
plane.color("blue")
plane.begin_fill()
plane.forward(200)
plane.left(90)
plane.forward(50)
plane.left(90)
plane.forward(200)
plane.left(90)
plane.forward(50)
plane.end_fill()

# Draw the wings
plane.penup()
plane.goto(-50, 50)
plane.pendown()
plane.begin_fill()
plane.color("red")
plane.forward(100)
plane.left(90)
plane.forward(50)
plane.left(90)
plane.forward(100)
plane.left(90)
plane.forward(50)
plane.end_fill()

# Draw the tail
plane.penup()
plane.goto(100, 0)
plane.pendown()
plane.begin_fill()
plane.color("green")
plane.forward(50)
plane.left(120)
plane.forward(100)
plane.left(120)
plane.forward(100)
plane.left(120)
plane.forward(50)
plane.end_fill()

# Draw the windows
plane.penup()
plane.goto(-80, 10)
plane.pendown()
plane.color("black")
for _ in range(2):
    plane.begin_fill()
    plane.circle(10)
    plane.end_fill()
    plane.penup()
    plane.forward(80)
    plane.pendown()

# Hide the turtle
plane.hideturtle()

# Keep the window open
turtle.done()
```

```python
import turtle

# Set up the screen
screen = turtle.Screen()
screen.setup(width=800, height=600)
screen.bgcolor("skyblue")

# Create a turtle for drawing the plane
plane = turtle.Turtle()
plane.speed(0)  # Set the drawing speed to fastest

# Function to draw the body of the plane
def draw_plane_body():
    plane.penup()
    plane.goto(-50, 0)
    plane.pendown()
    plane.color("blue")
    plane.begin_fill()
    plane.forward(100)
    plane.left(90)
    plane.forward(20)
    plane.left(90)
    plane.forward(100)
    plane.left(90)
    plane.forward(20)
    plane.end_fill()

# Function to draw the wings of the plane
def draw_wings():
    plane.penup()
    plane.goto(-30, 20)
    plane.pendown()
    plane.color("red")
    plane.begin_fill()
    plane.goto(-10, 20)
    plane.goto(-20, 50)
    plane.goto(-30, 20)
    plane.end_fill()

    plane.penup()
    plane.goto(0, 20)
    plane.pendown()
    plane.begin_fill()
    plane.goto(20, 20)
    plane.goto(10, 50)
    plane.goto(0, 20)
    plane.end_fill()

# Function to draw the tail of the plane
def draw_tail():
    plane.penup()
    plane.goto(50, 0)
    plane.pendown()
    plane.color("green")
    plane.begin_fill()
    plane.goto(60, 0)
    plane.goto(60, 20)
    plane.goto(50, 40)
    plane.goto(50, 0)
    plane.end_fill()

# Draw the plane
draw_plane_body()
draw_wings()
draw_tail()

# Hide the turtle
plane.hideturtle()

# Keep the window open
screen.mainloop()
```

```python
from turtle import *

def draw_plane(x=0, y=0, dir='up', pln_color='red', bg_color='cyan'):
    pensize(2)
    speed(0)

    hideturtle()
    penup()
    goto(x-40, y-40)
    setheading(0)
    pendown()
    fillcolor(bg_color)
    begin_fill()
    for i in range(4):
        forward(80)
        left(90)
    end_fill()
    penup()
    home()

    fillcolor(pln_color)

    if(dir == 'up'):
        goto(x-4, y-30)
    elif(dir == 'down'):
        goto(x+4, y+30)
        right(180)
    elif(dir=='right'):
        goto(x-30, y+4)
        right(90)

    begin_fill()
    pendown()
    left(100)
    forward(20)
    right(10)
    forward(35)
    circle(-6, 180)
    forward(35)
    right(10)
    forward(20)
    end_fill()
    begin_fill()
    circle(-2, 180)
    end_fill()
    begin_fill()
    # setheading(195)
    left(115) #radder
    forward(15)
    right(105)
    forward(5)
    right(50)
    forward(16)
    penup()
    # setheading(0)
    right(40)
    forward(8)
    pendown()
    right(40)
    forward(16)
    right(50)
    forward(5)
    right(105)
    forward(15)
    end_fill()

    penup()

    # home()
    if(dir=='up'):
        goto(x,y)
        setheading(0)
    elif(dir=='down'):
        goto(x,y)
        setheading(180)
    elif(dir=='right'):
        goto(x,y)
        setheading(-90)

    begin_fill()
    forward(6)
    right(90)
    forward(3)
    pendown()
    left(80)
    forward(22)
    left(100)
    forward(8)
    left(70)
    forward(23)
    left(20)
    penup()
    forward(13)
    pendown()
    left(20)
    forward(23)
    left(70)
    forward(8)
    left(100)
    forward(22)
    right(10)


    end_fill()
```

```python
from turtle import *


def draw_plane2(x=0, y=0, dir='up'):
    def draw_rect(x=0, y=0, length=20, width=10, clr='blue'):
        nonlocal dir
        penup()
        if(dir=='up'):
            goto(x-width/2, y-length/2)
        elif(dir=='down'):
            goto(x+width/2, y+length/2)
            setheading(180)
        elif(dir=='right'):
            goto(x-length/2,y+width/2)
            setheading(270)
        fillcolor(clr)
        begin_fill()
        pendown()
        for i in range(2):
            forward(width)
            left(90)
            forward(length)
            left(90)
        end_fill()
        penup()

    def draw_wing(x=0, y=0, length=30, width=10, clr='light gray'):
        nonlocal dir

        penup()
        goto(x,y)
        if(dir=='up'):
            setheading(0)
        elif(dir=='down'):
            setheading(180)
        elif(dir=='right'):
            setheading(270)
        right(135)
        fillcolor(clr)
        begin_fill()
        pendown()
        forward(length)
        left(45)
        forward(width)
        left(120)
        forward(length)
        end_fill()
        penup()
        goto(x,y)
        if(dir=='up'):
            setheading(0)
        elif(dir=='down'):
            setheading(180)
        elif(dir=='right'):
            setheading(270)
        right(45)
        begin_fill()
        pendown()
        forward(length)
        right(45)
        forward(width)
        right(120)
        forward(length)
        end_fill()
        setheading(0)

    # pensize(2)
    # speed(0)

    #drawing the frame
    hideturtle()
    penup()
    goto(x-40, y-40)
    setheading(0)
    pendown()
    # fillcolor(bg_color)
    color('slate gray')
    if(dir=='up'):
        fillcolor('steelblue1')
    elif(dir=='right'):
        fillcolor('gray')
    elif(dir=='down'):
        fillcolor('sienna2')
    begin_fill()
    for i in range(4):
        forward(80)
        left(90)
    # color('black')
    end_fill()
    penup()
    if(dir=='right'):
        goto(x-36,y+33)
        setheading(0)
        pensize(3)
        color('white')
        for i in range(3):
            pendown()
            forward(20)
            penup()
            forward(6)
        penup()
        goto(x-36,y-33)
        setheading(0)
        for i in range(3):
            pendown()
            forward(20)
            penup()
            forward(6)
    penup()
    pensize(1)
    color('black')
    home()

    #drawing the plane
    if(dir=='up'):
        draw_rect(x+15, y+7, 10, 7, 'white')
        draw_rect(x-15, y+7,10,7, 'white')
        draw_wing(x,y+22,36)
        draw_rect(x,y+10, 40,10, 'white')
        draw_wing(x,y-14, 21, 5)
        draw_rect(x,y-18,16,10,'red')
        penup()
    elif(dir=='down'):
        draw_rect(x-15, y-7, 10, 7, 'white')
        draw_rect(x+15, y-7,10,7, 'white')
        draw_wing(x,y-22,36)
        draw_rect(x,y-10, 40,10, 'white')
        draw_wing(x,y+10, 20, 5)
        draw_rect(x,y+18,16,10,'red')
        penup()
    elif(dir=='right'):
        draw_rect(x+7, y+15, 10, 7, 'white')
        draw_rect(x+7, y-15,10,7, 'white')
        draw_wing(x+22,y,36)
        draw_rect(x+10,y, 40,10, 'white')
        draw_wing(x-10,y, 20, 5)
        draw_rect(x-18,y,16,10,'red')
        penup()

    if(dir=='up'):
        goto(x+5,y+30)
        setheading(0)
    elif(dir=='down'):
        goto(x-5,y-30)
        setheading(180)
    elif(dir=='right'):
        goto(x+30,y-5)
        setheading(270)
    fillcolor('white')
    left(90)
    pendown()
    begin_fill()
    circle(5,180)
    end_fill()

    penup()
    if(dir=='up'):
        goto(x+5,y-26)
    elif(dir=='down'):
        goto(x-5,y+26)
        setheading(90)
    elif(dir=='right'):
        goto(x-26,y-5)
        setheading(-180)
    fillcolor('red')
    pendown()
    begin_fill()
    circle(-5,180)
    end_fill()
    # color('gray')

    #drawing the tail
    if(dir=='up'):
        setheading(0)
        draw_rect(x,y-24, 20,4,'red')
    elif(dir=='down'):
        setheading(180)
        draw_rect(x,y+24, 20,4,'red')
    elif(dir=='right'):
        setheading(270)
        draw_rect(x-24,y, 20,4,'red')

clearscreen()
speed(0)
pensize(0)

penup()
goto(100,-80)
pendown()
write('Rj Avro Biswas',font=["Arial",12,"normal"])
draw_plane2(0,0,'up')
draw_plane2(80,80,'down')
draw_plane2(160,-160,'right')
draw_plane2(160,80)

penup()
goto(0,-160)
pendown()
write('Rotation',font=["Arial",12,"bold"])
draw_plane2(-160,-160, 'right')

# draw_plane2(-80,-80,'down')
```
