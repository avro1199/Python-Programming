```python
import numpy as np
import matplotlib.pyplot as plt

def B(x, R, I):
    mu_0 = 4*np.pi*1e-7
    if(x<R):
        return (mu_0*I)/(2*np.pi*R)
    else:
        return (mu_0*I)/(2*np.pi*x)

x_data = np.linspace(0, 10, 1000)

plt.figure(1, dpi=240)
plt.plot(x_data, B(3, 6, x_data))
plt.show()
```
